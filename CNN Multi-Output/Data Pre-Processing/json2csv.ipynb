```python
import pandas as pd
import json
import os
from tqdm import tqdm

folder_path = '/home/lab38/images/딸기_증강_라벨'

dirListing = os.listdir(folder_path)
result = pd.DataFrame()

for i,fileName in enumerate(tqdm(dirListing)):
    address = os.path.join(folder_path, fileName)   
    with open(address, 'r') as f:
        data = json.load(f)
        df1 = pd.json_normalize(data["description"])
        df2 = pd.json_normalize(data["annotations"])
        df3 = pd.concat([df1,df2], axis = 1, sort=True)
        result = result.append(df3)

result.drop(['date', 'worker', 'height', 'width', 'task', 'type', 'region','crop', 'risk'], axis=1, inplace=True)        
result.to_csv("/home/lab38/images/csv/딸기_증강_라벨.csv", index=False)

display(result.head())
```
